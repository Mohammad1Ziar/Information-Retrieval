```python
documents = {
    "doc1": "Machine learning and data science are fields that utilize data to generate insights, build predictive models, and automate processes.",
    "doc2": "Data mining techniques are essential in data science and analytics for discovering patterns and relationships in large datasets.",
    "doc3": "Natural language processing is a subfield of artificial intelligence that focuses on understanding and generating human language.",
    "doc4": "Information retrieval deals with the search and extraction of relevant information from large collections, often using machine learning techniques.",
    "doc5": "Predictive analytics in data science often leverages machine learning and data mining to forecast future trends based on historical data."
}

def preprocess(doc):
    return doc.lower().replace("\n", " ").replace("(", "").replace(")", "").replace(",", "").replace(".", "").split()

def unigram_model(doc):
    tokens = preprocess(doc)
    word_counts = Counter(tokens)
    total_words = len(tokens)
    unigram_probabilities = {word: count / total_words for word, count in word_counts.items()}
    return unigram_probabilities

unigram_models = {doc_id: unigram_model(doc) for doc_id, doc in documents.items()}

for doc_id, model in unigram_models.items():
    print(f"Unigram Model for {doc_id}:")
    for word, prob in sorted(model.items(), key=lambda x: x[1], reverse=True):
        print(f"{word}: {prob:.6f}")
    print()
```

```python
def jelinek_mercer_smoothing(word, doc_model, background_model, lambda_):
    prob_doc = doc_model.get(word, 1e-3) 
    
    prob_bg = background_model.get(word, 1e-2) 
    

    return lambda_ * prob_doc + (1 - lambda_) * prob_bg

def calculate_query_probability(query, doc_model, background_model, lambda_):
    prob = 1.0
    for word in query:
        prob *= jelinek_mercer_smoothing(word, doc_model, background_model, lambda_)
    return prob

verbose_query = ["what", "are", "the", "techniques", "used", "in", "data", "science", "and", "machine", "learning"]
keyword_query = ["data", "science", "techniques"]

background_model = {
    "and": 0.058190,
    "data": 0.038793,
    "of": 0.023707,
    "the": 0.023707,
    "to": 0.021552,
}

lambda_ = 0.5  

query_probs_doc1 = calculate_query_probability(verbose_query, unigram_models["doc1"], background_model, lambda_)
query_probs_doc2 = calculate_query_probability(verbose_query, unigram_models["doc2"], background_model, lambda_)
query_probs_doc3 = calculate_query_probability(verbose_query, unigram_models["doc3"], background_model, lambda_)
query_probs_doc4 = calculate_query_probability(verbose_query, unigram_models["doc4"], background_model, lambda_)
query_probs_doc5 = calculate_query_probability(verbose_query, unigram_models["doc5"], background_model, lambda_)

query_probs_doc11 = calculate_query_probability(keyword_query, unigram_models["doc1"], background_model, lambda_)
query_probs_doc22 = calculate_query_probability(keyword_query, unigram_models["doc2"], background_model, lambda_)
query_probs_doc33 = calculate_query_probability(keyword_query, unigram_models["doc3"], background_model, lambda_)
query_probs_doc44 = calculate_query_probability(keyword_query, unigram_models["doc4"], background_model, lambda_)
query_probs_doc55 = calculate_query_probability(keyword_query, unigram_models["doc5"], background_model, lambda_)


print(f"Verbose Query Probabilities (Lambda = {lambda_}):")
print(f"Doc1: {query_probs_doc1:.10e}")
print(f"Doc2: {query_probs_doc2:.10e}")
print(f"Doc3: {query_probs_doc3:.10e}")
print(f"Doc4: {query_probs_doc4:.10e}")
print(f"Doc5: {query_probs_doc5:.10e}")

print(f"keyword Query Probabilities (Lambda = {lambda_}):")
print(f"Doc1: {query_probs_doc11:.10e}")
print(f"Doc2: {query_probs_doc22:.10e}")
print(f"Doc3: {query_probs_doc33:.10e}")
print(f"Doc4: {query_probs_doc44:.10e}")
print(f"Doc5: {query_probs_doc55:.10e}")
```

```python
verbose_threshold = 2e-22
keyword_threshold = 2e-7

relevant_documents_verbose = {"doc1", "doc3", "doc4"}
relevant_documents_keyword = {"doc1", "doc3", "doc4"}

def calculate_precision(retrieved_docs, relevant_docs):
    retrieved_relevant = len(retrieved_docs.intersection(relevant_docs))
    precision = retrieved_relevant / len(retrieved_docs) if len(retrieved_docs) > 0 else 0
    return precision

def get_retrieved_documents(query_probabilities, threshold):
    return {doc_id for doc_id, prob in query_probabilities.items() if prob > threshold}

retrieved_docs_verbose = get_retrieved_documents({
    "doc1": query_probs_doc1,
    "doc2": query_probs_doc2,
    "doc3": query_probs_doc3,
    "doc4": query_probs_doc4,
    "doc5": query_probs_doc5
}, verbose_threshold)

retrieved_docs_keyword = get_retrieved_documents({
    "doc1": calculate_query_probability(keyword_query, unigram_models["doc1"], background_model, lambda_),
    "doc2": calculate_query_probability(keyword_query, unigram_models["doc2"], background_model, lambda_),
    "doc3": calculate_query_probability(keyword_query, unigram_models["doc3"], background_model, lambda_),
    "doc4": calculate_query_probability(keyword_query, unigram_models["doc4"], background_model, lambda_),
    "doc5": calculate_query_probability(keyword_query, unigram_models["doc5"], background_model, lambda_)
}, keyword_threshold)

precision_verbose = calculate_precision(retrieved_docs_verbose, relevant_documents_verbose)
precision_keyword = calculate_precision(retrieved_docs_keyword, relevant_documents_keyword)

print(f"Precision for Verbose Query (Lambda = {lambda_}): {precision_verbose:.4f}")
print(f"Precision for Keyword Query (Lambda = {lambda_}): {precision_keyword:.4f}")
```

```python
lambda_values = [i / 10 for i in range(11)]  

precision_verbose_values = []
precision_keyword_values = []

for lambda_ in lambda_values:
    retrieved_docs_verbose = get_retrieved_documents({
        "doc1": calculate_query_probability(verbose_query, unigram_models["doc1"], background_model, lambda_),
        "doc2": calculate_query_probability(verbose_query, unigram_models["doc2"], background_model, lambda_),
        "doc3": calculate_query_probability(verbose_query, unigram_models["doc3"], background_model, lambda_),
        "doc4": calculate_query_probability(verbose_query, unigram_models["doc4"], background_model, lambda_),
        "doc5": calculate_query_probability(verbose_query, unigram_models["doc5"], background_model, lambda_)
    }, verbose_threshold)

    retrieved_docs_keyword = get_retrieved_documents({
        "doc1": calculate_query_probability(keyword_query, unigram_models["doc1"], background_model, lambda_),
        "doc2": calculate_query_probability(keyword_query, unigram_models["doc2"], background_model, lambda_),
        "doc3": calculate_query_probability(keyword_query, unigram_models["doc3"], background_model, lambda_),
        "doc4": calculate_query_probability(keyword_query, unigram_models["doc4"], background_model, lambda_),
        "doc5": calculate_query_probability(keyword_query, unigram_models["doc5"], background_model, lambda_)
    }, keyword_threshold)

    precision_verbose = calculate_precision(retrieved_docs_verbose, relevant_documents_verbose)
    precision_keyword = calculate_precision(retrieved_docs_keyword, relevant_documents_keyword)

    precision_verbose_values.append(precision_verbose)
    precision_keyword_values.append(precision_keyword)

plt.figure(figsize=(10, 6))
plt.plot(lambda_values, precision_verbose_values, label='Verbose Query Precision', marker='o')
plt.plot(lambda_values, precision_keyword_values, label='Keyword Query Precision', marker='x')

plt.xlabel('Lambda')
plt.ylabel('Precision')
plt.title('Precision vs Lambda for Verbose and Keyword Queries')
plt.legend()

plt.grid(True)
plt.show()
```
