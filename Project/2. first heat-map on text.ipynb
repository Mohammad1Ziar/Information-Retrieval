```python
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

# Generate a sample multiline text
sample_text = """ 
as 
it it
to to to
po po po po 
li li li li li 
ka ka ka ka ka ka 
mz mz mz mz mz mz mz 
re re re re re re re re
qw qw qw qw qw qw qw qw qw 
az az az az az az az az az az .
mn mn mn mn mn mn mn mn mn mn mn 
gb gb gb gb gb gb gb gb gb gb gb gb gb
ax ax ax ax ax ax ax ax ax ax ax ax ax ax
ad ad ad ad ad ad ad ad ad ad ad ad ad ad ad
"""

# Tokenize the text by splitting it into words and counting frequency
words = sample_text.lower().split()
word_counts = Counter(words)

# Create a colormap for visualization based on word frequency
max_count = max(word_counts.values())
min_count = min(word_counts.values())

# Create a function to assign color based on frequency
def get_color(count, min_count, max_count):
    # More frequent words are colder (bluer), less frequent are warmer (redder)
    norm = (count - min_count) / (max_count - min_count)
    return (1 - norm, 0, norm)  # RGB: (Red intensity, 0, Blue intensity)

# Generate colors for each word
word_colors = {word: get_color(count, min_count, max_count) for word, count in word_counts.items()}

# Plotting the heatmap-like visualization
fig, ax = plt.subplots(figsize=(12, 6))

# Display the words with colors based on their frequency
x, y = 0.5, 0.5  # Starting coordinates for the first word

# Adjusting the code to print each sentence on a new line

# Tokenize the text by splitting it into sentences and then words
sentences = [sentence.strip() for sentence in sample_text.strip().split('\n') if sentence.strip()]
word_counts = Counter(words)

# Reset the initial position for printing words
x, y = 0.5, 0.9  # Starting coordinates for the first word

# Iterate over each sentence and print words line by line
for sentence in sentences:
    sentence_words = sentence.lower().split()
    for word in sentence_words:
        color = word_colors[word]
        ax.text(x, y, word, fontsize=12, ha='center', va='center', color=color, fontweight='bold')
        x += len(word) * 0.05 + 0.1  # Adjust position for next word based on word length
        if x > 15:  # Wrap the text if it goes too far on the x-axis
            x = 0.5
            y -= 0.1
    # Move to the next line for the next sentence
    x = 0.5
    y -= 0.1

# Adjust the plot aesthetics
ax.set_xlim(0, 4)
ax.set_ylim(0, 1)
ax.axis('off')
plt.title('Textual Heatmap Visualization: Word Frequency-Based Coloring')
plt.show()
```

```python
# Generate a new sample multiline text
sample_text2 = """
hi to all of the people that are present at this house and 
tonight is the only night that we are here to celebrate 
some work together and no one knew . some day a man came to 
my room and said something to me so I ran out because I was alone in
my room and I on no on no had had hadwanted to had luch because it was noon.
now I am alone my me knew self and no one is here to talk
so what should I do now. considering this work that I am not 
here together alone in alone knew me in alone
"""

# Tokenize the new text by splitting it into sentences and then words
sentences2 = [sentence.strip() for sentence in sample_text2.strip().split('\n') if sentence.strip()]
words2 = sample_text2.lower().split()
word_counts2 = Counter(words2)

# Create a colormap for visualization based on word frequency for the new text
max_count2 = max(word_counts2.values())
min_count2 = min(word_counts2.values())

# Create a function to assign color based on frequency for the new text
def get_color2(count, min_count, max_count):
    norm = (count - min_count) / (max_count - min_count)
    return (1 - norm, 0, norm)  # RGB: (Red intensity, 0, Blue intensity)

# Generate colors for each word in the new text
word_colors2 = {word: get_color2(count, min_count2, max_count2) for word, count in word_counts2.items()}

# Plotting the heatmap-like visualization with each word inside a rectangle for the new text
fig, ax = plt.subplots(figsize=(12, 6))

# Reset the initial position for printing words
x2, y2 = 0.5, 1.0  # Starting coordinates for the first word

# Iterate over each sentence and print words line by line with rectangles for the new text
for sentence in sentences2:
    sentence_words2 = sentence.lower().split()
    for word in sentence_words2:
        color = word_colors2[word]
        # Draw a rectangle with the background color based on the word frequency
        rect = plt.Rectangle((x2-0.05, y2-0.05), len(word) * 0.05 + 0.1, 0.1, color=color, alpha=0.6)
        ax.add_patch(rect)
        # Print the word inside the rectangle with a white color font
        ax.text(x2 + (len(word) * 0.05) / 2, y2, word, fontsize=12, ha='center', va='center', color='white', fontweight='bold')
        x2 += len(word) * 0.05 + 0.15  # Adjust position for the next word
        if x2 > 15:  # Wrap the text if it goes too far on the x-axis
            x2 = 0.5
            y2 -= 0.10
    # Move to the next line for the next sentence
    x2 = 0.5
    y2 -= 0.15

# Adjust the plot aesthetics
ax.set_xlim(0, 6)
ax.set_ylim(0, 1.1)
ax.axis('off')
plt.title('Textual Heatmap Visualization: Word Frequency-Based Coloring with Rectangles (New Text)')
plt.show()
```
