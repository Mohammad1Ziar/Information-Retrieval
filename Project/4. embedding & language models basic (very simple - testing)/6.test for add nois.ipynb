```python
import random

def add_noise_to_word(word):
    operations = ['delete', 'insert', 'swap']
    op = random.choice(operations)
    if len(word) < 2:  # Handle short words
        return word
    if op == 'delete':
        idx = random.randint(0, len(word) - 1)
        return word[:idx] + word[idx+1:]
    elif op == 'insert':
        idx = random.randint(0, len(word))
        return word[:idx] + random.choice('abcdefghijklmnopqrstuvwxyz') + word[idx:]
    elif op == 'swap':
        idx = random.randint(0, len(word) - 2)
        return word[:idx] + word[idx+1] + word[idx] + word[idx+2:]

print(add_noise_to_word("Im a student at isfahan University of technology"))
print(add_noise_to_word("hello world"))
print(add_noise_to_word("library"))
print(add_noise_to_word("book"))
print(add_noise_to_word("play game"))
print(add_noise_to_word("where are you?"))
print(add_noise_to_word("He cant control him self"))
print(add_noise_to_word("He's hungry."))
print(add_noise_to_word("He is home alone"))
print(add_noise_to_word("He was listening to the music"))
print(add_noise_to_word("I was noit there."))
print(add_noise_to_word("She's here"))
print(add_noise_to_word("We are going home."))

#it was just a test
```
