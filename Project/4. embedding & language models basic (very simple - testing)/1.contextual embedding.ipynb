```python
import pandas as pd

data = pd.read_csv('train.csv')

sampled_data = data.sample(n=2000, random_state=42)
```

```python
sentences = sampled_data['Title'].tolist()
labels = sampled_data['Class Index'].tolist()
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2", pad_token_id=50256)

augmented_sentences = []
for sentence in sentences:
    new_sentences = generator(
        sentence,
        max_length=40,
        num_return_sequences=1,
        temperature=0.7,
        top_p=0.9,
        truncation=True
    )
    for new_sentence in new_sentences:
        augmented_sentences.append(new_sentence['generated_text'])

print(augmented_sentences[:10])
```
