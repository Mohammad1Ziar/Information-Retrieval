```python
file_path = '/content/drive/MyDrive/IUT/IR/project/4. contextual embedding/train.csv'
```

```python
import pandas as pd

data = pd.read_csv(file_path)
print(data.head())
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2", pad_token_id=50256)

augmented_sentences = []
sampled_data = data.sample(n=1000, random_state=42)
sentences = sampled_data['Title'].tolist()

for sentence in sentences:
    new_sentences = generator(
        sentence,
        max_new_tokens=30,
        num_return_sequences=2,
        temperature=0.7,
        top_p=0.9,
        truncation=True
    )
    for new_sentence in new_sentences:
        augmented_sentences.append(new_sentence['generated_text'])

augmented_data = pd.DataFrame({
    'Original': sentences * 2,
    'Generated': augmented_sentences
})

augmented_data.to_csv('/content/drive/IUT/IR/project/4. contextual embedding/augmented_dataset.csv', index=False)

print("Dataset has been saved as 'augmented_dataset.csv'")
```
